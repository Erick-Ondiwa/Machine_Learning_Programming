# Line Ploting
Line plots are one of the most commonly used plots in data visualization. They help track changes over time or display trends across ordered data. A line plot is created by plotting data points along an axis and then connecting them with a line, making it easier to see overall trends, seasonal patterns, and fluctuations

# Practical Example: Plotting Monthly Sales Data
Imagine you are analyzing a companyâ€™s monthly sales data over a year to understand seasonal patterns, trends, and peaks.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data: Monthly sales figures (in thousands)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = [15, 18, 25, 30, 45, 50, 48, 42, 38, 35, 28, 20]

# Create the line plot
plt.plot(months, sales, color='blue', marker='.', linestyle='-', linewidth=2, markersize=6)

# Adding labels and title
plt.xlabel("Month")
plt.ylabel("Sales (in $1000s)")
plt.title("Monthly Sales Data for 2023")

# Customizing ticks and grid
plt.xticks(rotation=45)  # Rotate month labels for better readability
plt.grid(True, color='green', linestyle='--')           # Add grid for easier reading of data points

# Display the plot
plt.show()
```

```python
# Sample data
data = np.array([10, 20, 30, 40, 50, 60, 70])

# Define a 3-point averaging kernel
kernel = np.ones(3) / 3  # This creates [1/3, 1/3, 1/3]

# Calculate the moving average with 'same' mode to keep the original array length
moving_average = np.convolve(data, kernel, mode='same')

print("Original Data:", data)
print("Moving Average:", moving_average)
```

# Example: Adding a Trend Line and Shaded Area

```python
# Adding a trend line (moving average)
moving_avg = np.convolve(sales, np.ones(3)/3, mode='same')

# Plotting original sales data
plt.plot(months, sales, color='blue', marker='o', linestyle='-', linewidth=2, markersize=6, label="Monthly Sales")

# Adding a trend line (moving average)
plt.plot(months, moving_avg, color='red', linestyle='--', linewidth=2, label="3-Month Moving Average")

# Shading the area under the sales line
plt.fill_between(months, sales, color='skyblue', alpha=0.2)

# Labels, title, legend, and grid
plt.xlabel("Month")
plt.ylabel("Sales (in $1000s)")
plt.title("Monthly Sales Data with Moving Average")
plt.legend()
plt.grid(True)
plt.show()
```
