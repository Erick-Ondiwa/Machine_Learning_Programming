# Bar Charts
# Practical Example: Sales Data by Product Category
Suppose you have sales data for a store’s product categories. We’ll visualize the sales to easily compare the performance across categories.

```python
import matplotlib.pyplot as plt

# Data for the bar chart
categories = ["Electronics", "Clothing", "Furniture", "Toys", "Food", "Stationery"]
sales = [25, 15, 30, 10, 20, 5]

# Create the bar chart
plt.bar(categories, sales, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel("Product Category")
plt.ylabel("Sales (in $1000s)")
plt.title("Sales by Product Category")

plt.grid(True, color='green', linestyle='--')

# Customize x-ticks for readability
plt.xticks(rotation=45)

# Adding labels on top of each bar
for i, value in enumerate(sales):
    plt.text(i, value + 1, f"{value}", ha='center')  # Slightly above the top of the bar
# Display the chart
plt.show()
```

# Horizontal Bar Chart
A horizontal bar chart can be created using plt.barh() and is useful for lengthy category labels.

```python
plt.barh(categories, sales, color='lightgreen', edgecolor='black')
plt.xlabel("Sales (in $1000s)")
plt.ylabel("Product Category")
plt.title("Sales by Product Category")
plt.show()
```

# Grouped Bar Chart
If you want to compare multiple datasets, such as sales figures from two different years, a grouped bar chart is useful.

```python
import numpy as np

# Additional data for previous year's sales
sales_2022 = [20, 12, 28, 8, 18, 7]

# Set bar width and positions
bar_width = 0.35
index = np.arange(len(categories))

# Plot both sets of bars side by side
plt.bar(index, sales, width=bar_width, color='skyblue', label='2023')
plt.bar(index + bar_width, sales_2022, width=bar_width, color='salmon', label='2022')

# Labeling and formatting
plt.xlabel("Product Category")
plt.ylabel("Sales (in $1000s)")
plt.title("Sales by Product Category (2022 vs 2023)")
plt.xticks(index + bar_width / 2, categories, rotation=45)
plt.legend()

plt.show()
```
