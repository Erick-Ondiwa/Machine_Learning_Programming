# Creating Histograms
A histogram is a powerful tool in data visualization that shows the distribution of a dataset by dividing it into intervals, or "bins," and plotting the frequency of data points within each bin. Histograms are especially useful for understanding the shape, spread, and central tendency of data.

# Practical Example: Analyzing the Distribution of Student Exam Scores
Let's say you have a dataset of student exam scores, and you want to understand the distribution of these scores to identify how students performed overall.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: Exam scores of students
scores = [56, 78, 45, 67, 89, 53, 90, 77, 66, 82, 95, 88, 72, 60, 75, 69, 55, 85, 92, 47,
          72, 80, 68, 73, 84, 91, 65, 70, 82, 76, 94, 61, 58, 63, 66, 59, 83, 87, 64, 57]

# Create the histogram
plt.hist(scores, bins=10, color='skyblue', edgecolor='black', label="Exam Scores")

# Adding labels and title
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.title("Distribution of Student Exam Scores")

# Place the legend outside the plot area
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Display the histogram
plt.show()
```

```python
plt.hist(scores, bins=10, color='skyblue', edgecolor='black', label="Exam Scores", orientation='horizontal')
# This changes the orientation of the bars to horizontal

plt.show()
```

# Overlapping Histograms:

For comparison, you could overlay two histograms, such as scores from two different exams.

```python
# Sample additional data
scores_2 = [59, 62, 65, 68, 73, 77, 78, 81, 84, 86, 88, 90, 91, 92, 94, 96, 97, 98, 89, 83]

# Plot both histograms with transparency
plt.hist(scores, bins=10, color='skyblue', edgecolor='black', label="Exam 1 Scores", alpha=0.6)
plt.hist(scores_2, bins=10, color='salmon', edgecolor='black', label="Exam 2 Scores", alpha=0.6)

# Legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.title("Distribution of Student Exam Scores for Two Exams")
plt.show()
```
