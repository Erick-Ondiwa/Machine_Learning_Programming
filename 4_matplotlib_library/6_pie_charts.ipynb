# Creating Pie Charts
A pie chart is a circular chart divided into sectors, with each sector representing a portion of the whole. Itâ€™s commonly used to show percentages or proportional data, making it an ideal way to visualize how categories contribute to a total.

# Example: Market Share of Smartphone Brands
Suppose you have data on smartphone market shares of various brands, and you want to create a pie chart to visualize their relative popularity.

```python
import matplotlib.pyplot as plt

# Data for the pie chart
brands = ["Apple", "Samsung", "Xiaomi", "Oppo", "Others"]
market_share = [30, 25, 20, 15, 10]

# Create the pie chart
plt.pie(market_share, labels=brands, autopct='%1.1f%%', startangle=140)

# Add title
plt.title("Smartphone Market Share by Brand")

# Display the chart
plt.show()
```
