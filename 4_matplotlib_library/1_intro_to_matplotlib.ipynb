# Basic Introduction
Matplotlib is a powerful, widely-used Python library for creating static, interactive, and animated visualizations. It offers tools to make charts, graphs, and plots. Matplotlib can be used to create line plots, bar charts, scatter plots, histograms, and more, which help in analyzing and interpreting data visually.

The libraryâ€™s main component is pyplot, a module within Matplotlib, which provides a convenient MATLAB-like interface for plotting. You can import it as follows:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 6)
print(x)
```

```python
y = np.arange(2,11,2)
print(y)
```

```python
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")
plt.show()
```
