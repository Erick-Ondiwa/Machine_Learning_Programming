# Introduction
Matplotlib is a powerful, widely-used Python library for creating static, interactive, and animated visualizations. It offers tools to make charts, graphs, and plots. Matplotlib can be used to create line plots, bar charts, scatter plots, histograms, and more, which help in analyzing and interpreting data visually.

The libraryâ€™s main component is pyplot, a module within Matplotlib, which provides a convenient MATLAB-like interface for plotting. You can import it as follows:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 6)
print(x)
```

```python
y = np.arange(2,11,2)
print(y)
```

```python
plt.plot(x,y, 'r.-')
plt.xlabel("X-axis", fontsize=12, fontweight='bold')
plt.ylabel("Y-axis", fontsize=12, color='black')
plt.title("Basic Line Plot", fontsize=12, color='black')
plt.grid(True, color='green', linestyle='--')

plt.legend('x')
plt.show()
```

```python
gas_price_data = pd.read_csv('gas_prices.csv')
gas_price_data.head()
```

```python
for data in gas_price_data:
    print(data)
```

```python
plt.title('Gas Price Variation Over time', fontsize=12, fontweight="bold")
plt.ylabel("Price(USD)", fontsize=12, color='black', fontfamily='Arial')
plt.xlabel("Year", fontsize=12, color='black', fontfamily='Arial')


country_idx =['Canada', 'Japan', 'Germany', 'South Korea', 'USA']
for country in gas_price_data:
    if country in country_idx:
        plt.plot(gas_price_data.Year, gas_price_data[country], marker='*')
        plt.xticks(gas_price_data.Year[::2])

plt.legend()
plt.show()
# y_axis = gas_price_data['Italy']
# print(y_axis)
```

```python
plt.plot(gas_price_data.Year, gas_price_data.Italy, 'g.-', label='Italy')
plt.plot(gas_price_data.Year, gas_price_data.Japan, 'y.-', label='Japan')
plt.plot(gas_price_data.Year, gas_price_data.UK, 'b.-', label='UK')
plt.title('Gas Price Variation Over time', fontsize=12, fontweight="bold")
plt.ylabel("Price(USD)", fontsize=12, color='black', fontfamily='Arial')
plt.xlabel("Year", fontsize=12, color='black', fontfamily='Arial')

plt.xticks(gas_price_data.Year[::2], rotation=45)
plt.legend()
plt.grid(True, linestyle='--', color='green', linewidth=0.5)
plt.show()
```

```python
# Sample data
house_area = [1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400]
house_price = [300, 320, 340, 360, 380, 400, 420, 440, 460, 480]

# Create a scatter plot
plt.scatter(house_area, house_price, color='blue', linestyle='-', marker='.', s=100, alpha=0.6, edgecolor='black')

# Adding labels and title
plt.xlabel("House Area (sq ft)")
plt.ylabel("Price ($1000s)")
plt.title("Relationship between House Area and Price")

plt.show()
```
